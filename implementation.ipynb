```python
from sklearn.preprocessing import StandardScaler

# Standardize data
scaler_cancer = StandardScaler()
cancer_scaled = scaler_cancer.fit_transform(cancer_df)

scaler_boston = StandardScaler()
boston_scaled = scaler_boston.fit_transform(boston_df)

scaler_iris = StandardScaler()
iris_scaled = scaler_iris.fit_transform(iris_df)
```

```python
from sklearn.decomposition import PCA, FastICA, TruncatedSVD
from sklearn.feature_selection import SelectKBest, f_classif, f_regression, RFE
from sklearn.linear_model import LogisticRegression, LinearRegression

# 1. BREAST CANCER - Classification

# PCA
pca_cancer = PCA()
pca_cancer.fit(cancer_scaled)
explained_var_cancer = pca_cancer.explained_variance_ratio_
cumsum_var_cancer = np.cumsum(explained_var_cancer)

n_components_95_cancer = np.argmax(cumsum_var_cancer >= 0.95) + 1
print(f"\nPCA: Components for 95% variance: {n_components_95_cancer}")
pca_cancer_final = PCA(n_components=n_components_95_cancer)
cancer_pca = pca_cancer_final.fit_transform(cancer_scaled)
print(f"PCA transformed shape: {cancer_pca.shape}")

# ICA
ica_cancer = FastICA(n_components=n_components_95_cancer, random_state=42)
cancer_ica = ica_cancer.fit_transform(cancer_scaled)
print(f"\nICA transformed shape: {cancer_ica.shape}")

# SVD
svd_cancer = TruncatedSVD(n_components=n_components_95_cancer, random_state=42)
cancer_svd = svd_cancer.fit_transform(cancer_scaled)
print(f"\nSVD transformed shape: {cancer_svd.shape}")
print(f"SVD explained variance ratio: {svd_cancer.explained_variance_ratio_.sum():.3f}")

# SelectKBest
selector_cancer = SelectKBest(score_func=f_classif, k=10)
cancer_selected = selector_cancer.fit_transform(cancer_df, cancer_target)
print(f"\nSelectKBest transformed shape: {cancer_selected.shape}")

# RFE
rfe_cancer = RFE(estimator=LogisticRegression(max_iter=10000), n_features_to_select=10)
cancer_rfe = rfe_cancer.fit_transform(cancer_df, cancer_target)
print(f"\nRFE transformed shape: {cancer_rfe.shape}")
```

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))
axes[0].plot(range(1, len(explained_var_cancer)+1), cumsum_var_cancer, 'bo-')
axes[0].axhline(y=0.95, color='r', linestyle='--', label='95% threshold')
axes[0].axvline(x=n_components_95_cancer, color='g', linestyle='--', 
                label=f'n={n_components_95_cancer}')
axes[0].set_xlabel('Number of Components')
axes[0].set_ylabel('Cumulative Explained Variance')
axes[0].set_title('Breast Cancer - PCA Explained Variance')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

scores_cancer = selector_cancer.scores_
axes[1].barh(range(len(scores_cancer)), scores_cancer)
axes[1].set_yticks(range(len(scores_cancer)))
axes[1].set_yticklabels(cancer_df.columns, fontsize=7)
axes[1].set_xlabel('F-score')
axes[1].set_title('Breast Cancer - Feature Importance (f_classif)')
axes[1].grid(True, alpha=0.3, axis='x')

plt.tight_layout()
plt.savefig('cancer_feature_extraction.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
# 2. BOSTON HOUSING - Regression

# PCA
pca_boston = PCA()
pca_boston.fit(boston_scaled)
explained_var_boston = pca_boston.explained_variance_ratio_
cumsum_var_boston = np.cumsum(explained_var_boston)
n_components_95_boston = np.argmax(cumsum_var_boston >= 0.95) + 1
print(f"\nPCA: Components for 95% variance: {n_components_95_boston}")
pca_boston_final = PCA(n_components=n_components_95_boston)
boston_pca = pca_boston_final.fit_transform(boston_scaled)
print(f"PCA transformed shape: {boston_pca.shape}")

# ICA
ica_boston = FastICA(n_components=n_components_95_boston, random_state=42)
boston_ica = ica_boston.fit_transform(boston_scaled)
print(f"\nICA transformed shape: {boston_ica.shape}")

# SVD
svd_boston = TruncatedSVD(n_components=n_components_95_boston, random_state=42)
boston_svd = svd_boston.fit_transform(boston_scaled)
print(f"\nSVD transformed shape: {boston_svd.shape}")
print(f"SVD explained variance ratio: {svd_boston.explained_variance_ratio_.sum():.3f}")

# SelectKBest
selector_boston = SelectKBest(score_func=f_regression, k=6)
boston_selected = selector_boston.fit_transform(boston_df, boston_target)
print(f"\nSelectKBest transformed shape: {boston_selected.shape}")

# RFE
rfe_boston = RFE(estimator=LinearRegression(), n_features_to_select=6)
boston_rfe = rfe_boston.fit_transform(boston_df, boston_target)
print(f"\nRFE transformed shape: {boston_rfe.shape}")
```

```python
from sklearn.linear_model import LinearRegression, SGDRegressor
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
import numpy as np

# Split indices first
indices_train, indices_test = train_test_split(
    np.arange(len(boston_target)),
    test_size=0.2, 
    random_state=42
)

# Apply to all datasets (features are already scaled)
X_boston_orig_train = boston_scaled[indices_train]
X_boston_orig_test = boston_scaled[indices_test]
X_boston_pca_train = boston_pca[indices_train]
X_boston_pca_test = boston_pca[indices_test]
X_boston_sel_train = boston_rfe[indices_train]
X_boston_sel_test = boston_rfe[indices_test]

# Split target - NO SCALING
y_boston_train = boston_target[indices_train]
y_boston_test = boston_target[indices_test]
```

```python
print("\n--- Model 1: Linear Regression (Analytical Optimization) ---")

lr_orig = LinearRegression()
lr_orig.fit(X_boston_orig_train, y_boston_train)
lr_orig_pred = lr_orig.predict(X_boston_orig_test)
lr_orig_mse = mean_squared_error(y_boston_test, lr_orig_pred)
lr_orig_train_mse = mean_squared_error(y_boston_train, lr_orig.predict(X_boston_orig_train))

lr_pca = LinearRegression()
lr_pca.fit(X_boston_pca_train, y_boston_train)
lr_pca_pred = lr_pca.predict(X_boston_pca_test)
lr_pca_mse = mean_squared_error(y_boston_test, lr_pca_pred)
lr_pca_train_mse = mean_squared_error(y_boston_train, lr_pca.predict(X_boston_pca_train))

print(f"\nOriginal Data:")
print(f"  Train MSE: {lr_orig_train_mse:.4f}")
print(f"  Test MSE: {lr_orig_mse:.4f}")
print(f"  Coefficient range: [{lr_orig.coef_.min():.4f}, {lr_orig.coef_.max():.4f}]")
print(f"  Coefficient std: {lr_orig.coef_.std():.4f}")

print(f"\nPCA Data:")
print(f"  Train MSE: {lr_pca_train_mse:.4f}")
print(f"  Test MSE: {lr_pca_mse:.4f}")
print(f"  Coefficient range: [{lr_pca.coef_.min():.4f}, {lr_pca.coef_.max():.4f}]")
print(f"  Coefficient std: {lr_pca.coef_.std():.4f}")
```

```python
print("\n--- Model 2: SGDRegressor (Derivative-Based Optimization) ---")

target_scale = np.std(y_boston_train)
eta0_value = 0.001 / target_scale

sgd_orig = SGDRegressor(max_iter=10000, eta0=eta0_value, random_state=42)
sgd_orig.fit(X_boston_orig_train, y_boston_train)
sgd_orig_pred = sgd_orig.predict(X_boston_orig_test)
sgd_orig_mse = mean_squared_error(y_boston_test, sgd_orig_pred)
sgd_orig_train_mse = mean_squared_error(y_boston_train, sgd_orig.predict(X_boston_orig_train))

sgd_pca = SGDRegressor(max_iter=10000, eta0=eta0_value, random_state=42)
sgd_pca.fit(X_boston_pca_train, y_boston_train)
sgd_pca_pred = sgd_pca.predict(X_boston_pca_test)
sgd_pca_mse = mean_squared_error(y_boston_test, sgd_pca_pred)
sgd_pca_train_mse = mean_squared_error(y_boston_train, sgd_pca.predict(X_boston_pca_train))

print(f"\nOriginal Data:")
print(f"  Train MSE: {sgd_orig_train_mse:.4f}")
print(f"  Test MSE: {sgd_orig_mse:.4f}")
print(f"  Iterations: {sgd_orig.n_iter_}")

print(f"\nPCA Data:")
print(f"  Train MSE: {sgd_pca_train_mse:.4f}")
print(f"  Test MSE: {sgd_pca_mse:.4f}")
print(f"  Iterations: {sgd_pca.n_iter_}")
```

```python
n_epochs = 25
losses_orig, losses_pca = [], []

sgd_orig_vis = SGDRegressor(max_iter=10000, eta0=eta0_value, random_state=42)
sgd_pca_vis = SGDRegressor(max_iter=10000, eta0=eta0_value, random_state=42)

for _ in range(n_epochs):
    sgd_orig_vis.partial_fit(X_boston_orig_train, y_boston_train)
    sgd_pca_vis.partial_fit(X_boston_pca_train, y_boston_train)
    
    losses_orig.append(mean_squared_error(y_boston_train, sgd_orig_vis.predict(X_boston_orig_train)))
    losses_pca.append(mean_squared_error(y_boston_train, sgd_pca_vis.predict(X_boston_pca_train)))

plt.figure(figsize=(8, 5))
plt.plot(losses_orig, label='Original Data')
plt.plot(losses_pca, label='PCA Data')
plt.xlabel('Iteration')
plt.ylabel('Training MSE')
plt.title('SGDRegressor Convergence Comparison')
plt.legend()
plt.grid(True)
plt.savefig('regression_convergence.png', dpi=150, bbox_inches='tight')
plt.show()
```

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
import time

# PART B - CLUSTERING PROBLEM (Iris)

print("\n--- K-Means Clustering (EM-Based Optimization) ---")

# Original Data
start_time = time.time()
kmeans_orig = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_orig.fit(iris_scaled)
time_orig = time.time() - start_time
silhouette_orig = silhouette_score(iris_scaled, kmeans_orig.labels_)

# PCA Data
start_time = time.time()
kmeans_pca = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans_pca.fit(iris_pca)
time_pca = time.time() - start_time
silhouette_pca = silhouette_score(iris_pca, kmeans_pca.labels_)

print(f"\nOriginal Data:")
print(f"  Inertia: {kmeans_orig.inertia_:.4f}")
print(f"  Silhouette Score: {silhouette_orig:.4f}")
print(f"  Iterations: {kmeans_orig.n_iter_}")
print(f"  Time: {time_orig:.4f}s")

print(f"\nPCA Data:")
print(f"  Inertia: {kmeans_pca.inertia_:.4f}")
print(f"  Silhouette Score: {silhouette_pca:.4f}")
print(f"  Iterations: {kmeans_pca.n_iter_}")
print(f"  Time: {time_pca:.4f}s")
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

# PART C - CLASSIFICATION PROBLEM (Breast Cancer)

# Split data for classification
indices_train, indices_test = train_test_split(
    np.arange(len(cancer_target)),
    test_size=0.2, 
    random_state=42
)

# Apply to all datasets (features are already scaled)
X_cancer_orig_train = cancer_scaled[indices_train]
X_cancer_orig_test = cancer_scaled[indices_test]
X_cancer_pca_train = cancer_pca[indices_train]
X_cancer_pca_test = cancer_pca[indices_test]
X_cancer_sel_train = cancer_rfe[indices_train]
X_cancer_sel_test = cancer_rfe[indices_test]

# Split target - NO SCALING
y_cancer_train = cancer_target[indices_train]
y_cancer_test = cancer_target[indices_test]
```

```python
print("\n--- Model 1: K-Nearest Neighbors (Distance-Based) ---")

knn_orig = KNeighborsClassifier(n_neighbors=5)
knn_orig.fit(X_cancer_orig_train, y_cancer_train)
start_time = time.time()
knn_orig_pred = knn_orig.predict(X_cancer_orig_test)
time_knn_orig = time.time() - start_time
acc_knn_orig = accuracy_score(y_cancer_test, knn_orig_pred)

knn_pca = KNeighborsClassifier(n_neighbors=5)
knn_pca.fit(X_cancer_pca_train, y_cancer_train)
start_time = time.time()
knn_pca_pred = knn_pca.predict(X_cancer_pca_test)
time_knn_pca = time.time() - start_time
acc_knn_pca = accuracy_score(y_cancer_test, knn_pca_pred)

print(f"\nOriginal Data:")
print(f"  Accuracy: {acc_knn_orig:.4f}")
print(f"  Prediction Time: {time_knn_orig:.6f}s")

print(f"\nPCA Data:")
print(f"  Accuracy: {acc_knn_pca:.4f}")
print(f"  Prediction Time: {time_knn_pca:.6f}s")
```

```python
print("\n--- Model 2: Random Forest (Tree-Based Ensemble) ---")

rf_orig = RandomForestClassifier(n_estimators=100, random_state=42)
rf_orig.fit(X_cancer_orig_train, y_cancer_train)
rf_orig_pred = rf_orig.predict(X_cancer_orig_test)
acc_rf_orig = accuracy_score(y_cancer_test, rf_orig_pred)

rf_pca = RandomForestClassifier(n_estimators=100, random_state=42)
rf_pca.fit(X_cancer_pca_train, y_cancer_train)
rf_pca_pred = rf_pca.predict(X_cancer_pca_test)
acc_rf_pca = accuracy_score(y_cancer_test, rf_pca_pred)

rf_sel = RandomForestClassifier(n_estimators=100, random_state=42)
rf_sel.fit(X_cancer_sel_train, y_cancer_train)
rf_sel_pred = rf_sel.predict(X_cancer_sel_test)
acc_rf_sel = accuracy_score(y_cancer_test, rf_sel_pred)


print(f"\nOriginal Data:")
print(f"  Accuracy: {acc_rf_orig:.4f}")

print(f"\nPCA Data:")
print(f"  Accuracy: {acc_rf_pca:.4f}")

print(f"\nRFE Data:")
print(f"  Accuracy: {acc_rf_sel:.4f}")
```
